```python
grades = np.loadtxt('exam_grades.csv', #the file name
                  delimiter=',', #indicate the delimiter
                  skiprows=1) #skip the header

#verify the data was imported correctly
print(grades[:5,]) #show the first five rows of the data
```

```python
print("The total number of columns is: ", grades.shape[0])
print("The total number of observations is: ", grades.shape[1])

course_grades = grades[0:233, 4]
print(course_grades[:5,]) #show the first five rows of the data

print('The maximum value is: ', course_grades.max())
print('The minimum value is: ', course_grades.min())
```

```python
course_years = grades[0:233, 0]
print(course_years)
no_duplicate = list()
```

```python
#filter the array to no duplicates 
for x in course_years:
    if x not in no_duplicate:
        no_duplicate.append(x)
print(no_duplicate)
```

```python
#count number of students in each year 
num_2000 = 0; 
num_2001 = 0; 
num_2002 = 0; 
num_2003 = 0; 
for x in course_years: 
    if x == 2000: 
        num_2000 += 1 
    elif x == 2001:
        num_2001 += 1 
    elif x == 2002: 
        num_2002 += 1
    elif x == 2003:
        num_2003 += 1

print("The number of students in 2000: ", num_2000)
print("The number of students in 2001: ", num_2001)
print("The number of students in 2002: ", num_2002)
print("The number of students in 2003: ", num_2003)
```
