## 1. Lists

**Here are two ways to define an empty list**

### Concatenating Lists 
We can merge (i.e. **concatenate**) lists using the **+ operator** which will create a new list that contains the elements of both lists. 

**Note**: this is the reason it is important to know the python types that you are using because the **+ operator** works differently with lists than it does numbers. 

````python
student_names_1 = ['Jane', 'John', 'Sue']
student_names_2 = ['George', 'Judy', 'Elroy']

student_names = student_names_1 + student_names_2
````

````python
print(student_names)
````

### Slicing a List
Slicing is an operation that copies a subset of the elements from a list and uses it to create a new list. When slicing lists, you can provide the start and end indices to indicate the range of values. 

**The example below will slice the first two student names from the list; 0 is the starting index and 2 is the ending index**. When you run the cell below, you will notice that the element at index 2 is not displayed. The reason for this is that the ending index is not inclusive.

````python
#Slice the first two students from the list
student_names[0:2]
````

We can also slice a list by specifying only a starting index, or an ending index.

#### Specifying a Slice with Only a Starting Index

````python
#This is a slice starting from the 4th element of the list to the end.
student_names[3:]
````

#### Specifying a Slice with Only an Ending Index

````python
#Slice from the beginning of the list and stopping before the 3rd element.
student_names[:3] #the starting index 0 is assumed
````

#### Specifying a Slice with No Indices

````python
student_names[:] #retrieve all elements
````

### Iterate through a List using a For loop
When working with lists, you'll often need to iterate over its elements and either display the contents or perform a calculation using each element. You can use a for loop to iterate over the sequence of elements in a list.

Here is an example:

````python
#list elements can be accessed using their indices. We can display each element and its index using the following
for i in range(len(student_names)):  
    print(i, ":", student_names[i])
````

````python
#Here is an alternative approach to the loop above which displays each element (without its index)
for student in student_names:  
    print(student)
````

However, we can pass an argument e.g. `sort(reverse = True)` to sort the list in descending order.

The example below shows how to sort a list in ascending and descending order.

````python
#Sorting a list in ascending order
student_names.sort()
````

````python
print(student_names)
````

````python
#Sorting a list in descending order
student_names.sort(reverse=True)
````

## 2. Dictionaries
A dictionary is a data structure that stores data using a key-value pair format. It is similar to a typical dictionary that maps words to definitions. 


### How to Create a Dictionary
Dictionaries are created using the brace notation `{}`, and it can either be 1) empty or 2) initialized with comma separated key-value pairs that are enclosed inside the braces. 


**Here are two ways to define a dictionary**

````python
#create an empty dictionary
student_dict = {}

#create a dictionary with the names of students and initialize it with values (i.e. their respective grades)
student_dict = {'jane': 97, 'john': 89, 'mary': 86}
````

### How to Retrieve the Keys and Values
The key-value pair format helps to keep the data organized. However, there will be times that you will need to retrieve only the keys or the values in the dictionary. There are two methods that allow this: `keys()` and `values()`.

The example below demonstrates their use:

````python
student_dict.keys() #return the keys of a dictionary
````

````python
student_dict.values() #return the values of a dictionary
````

The result from `keys()` and `values()` returns a type called `dict_keys()` and `dict_values()` respectively. These can be converted to a list as follows:

````python

list(student_dict.keys()) #converts the keys to a list
````

````python
list(student_dict.values()) #converts the values to a list
````

**The elements in the list of keys and values can now be manipulated using standard list operations.**

### Accessing and Updating the Value Associated with a Key
The items in the dictionary can be accessed using its respective key; knowing the key allows you to unlock the value. Here is an example that shows how to obtain a student's grade, based on her name -- which is the key:

````python
student_dict['jane'] #retrieve the value associated with the key 'jane'
````

We can also update the associated value for a key as follows

````python
student_dict['jane'] = 92 #updates the value associated with the key 'jane'
````

````python
student_dict
````

### Adding a New Key–Value Pair
You can add a new item to a list by creating a new key and assigning a value to it. If the key exists, the value will be updated. However, if the key doesn't exist, it will create a new key-value pair.

````python
student_dict['calvin'] = 95 #since the key is non-existent, the calvin-95 pair will be added
````

### Accessing a Nonexistent Key
If you try to access a key that doesnt exist, an error will be displayed. For example:

````python
#student_dict['kelsey']
````

To avoid errors, use the built-in dictionary method `get()` to return the value of a key, or return `None` if the key is not found.

````python
student_dict.get('kelsey')
````

Alternatively, you can check whether a dictionary contains a specific key using the `in` operator. For example:

````python
#example 1
'kelsey' in student_dict
````

````python
#example 2
'calvin' in student_dict
````

### Iterate through a Dictionary

Similar to lists, there will be times that you will need to iterate through dictionary items, keys and/or values. You can use a for loop to iterate over the items in this collection.

Here are a few examples:

````python
#iterate through dictionary items()
for student in student_dict.items():
    print(student) #this prints each key-value pair as a tuple
````

````python
#iterate through keys()
for student in student_dict.keys():
    print(student)
````

````python
#iterate through values()
for grade in student_dict.values():
    print(grade)
````
